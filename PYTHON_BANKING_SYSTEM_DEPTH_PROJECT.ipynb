```python
import random
import getpass
class Bank_Account:
    def __init__(self,account_number,name,age,password,balance=0):
        self.account_number=account_number
        self.name=name
        self.age=age
        self.password=password
        self.balance=balance
    def deposit(self,amount):
        if amount>=0:
            self.balance+=amount
            print(f"The deposited amount is ${amount} and the balance is ${self.balance}")
        else:
            print('Please enter a valid amount')
    def withdrawl(self,amount):
        if amount>=0 and amount<=self.balance:
            self.balance-=amount
            print(f"The amount withdrawed is {amount} and the balance is {self.balance}")
        else:
            print("Please enter a valid amount")
    def get_balance(self):
        return self.balance
    def account_details(self):
        return (self.account_number,self,name,self.age,self.password,self.balance)
def create_account(accounts):
    account_number=random.randint(10**11,10**12-1)
    name=input("Enter your name : ")
    age=int(input("Enter your age : "))
    password=getpass.getpass("Enter your password: ")
    balance=int(input("Enter initial balnce: "))
    new_account=Bank_Account(account_number,name,age,password,balance)
    accounts.append(new_account)
    print(f"Account successfully created ! Your account number is {account_number}")
def enter_existing_accounts(accounts):
    account_number=int(input("Enter your account number : "))
    password=getpass.getpass("Enter your account password : ")
    for account in accounts:
        if account.account_number==account_number and account.password==password:
            return account
    print("Invalid account_number or password")
    None
        
def display_all_accounts(self):
    print("All accounts: ")
    for account in accounts:
        print(f"Account_number: {account.account_number},Name : {account.name},Age:{account.age},Password:{account.password},Balance:{account.balance}")
                   
accounts=[]
while True:
    print("1.Create a new account")
    print("2.Enter an existing acccount")
    print("3.Display all accounts")
    print("4.Exit")
    choice=int(input("Enter Your choices 1/2/3/4 : "))
    if choice ==1:
        create_account(accounts)
    elif choice==2:
        existing_account= enter_existing_accounts(accounts)
        if existing_account:
            print("1.Deposit money: ")
            print("2.Withdraw money: ")
            print("3.Get balance: ")
            print("4. Remove account:")
            print("5. Exit Existing account")
            user_choice=input("Enter your choice: ")
            if user_choice=='1':
                amount=float(input("Enter the amount too be deposited :"))
                existing_account.deposit(amount)
            elif user_choice=='2':
                amount=float(input("Enter the amount to be withdrawn: "))
                existing_account.withdrawl(amount)
            elif user_choice=='3':
                existing_account.get_balance()
            elif user_choice=='4':
                accounts.remove(existing_account)
                print("Account removed successfully!")
                break
            elif user_choice=='5':
                 break
            else:
                print("Please enter a valid choice.")
    elif choice==3:
        display_all_accounts(accounts)
    elif choice==4:
         break
    else:
        print("Please enter a valid choice")
                    
```
